```python
# Set plot style
sns.set(style="whitegrid")

# Create a 2x2 subplot for key variables over time
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Temperature trends
sns.lineplot(data=climate_df, x="year", y="avg_mean_temp", ax=axs[0, 0], label="Mean Temp")
sns.lineplot(data=climate_df, x="year", y="avg_min_temp", ax=axs[0, 0], label="Min Temp")
sns.lineplot(data=climate_df, x="year", y="avg_max_temp", ax=axs[0, 0], label="Max Temp")
axs[0, 0].set_title("Temperature Trends Over Time")
axs[0, 0].legend()

# Humidity and rainfall
sns.lineplot(data=climate_df, x="year", y="relative_humidity", ax=axs[0, 1], label="Humidity")
sns.lineplot(data=climate_df, x="year", y="annual_rainfall", ax=axs[0, 1], label="Rainfall")
axs[0, 1].set_title("Humidity and Rainfall Over Time")
axs[0, 1].legend()

# Agricultural indicators
sns.lineplot(data=climate_df, x="year", y="agri_land_area", ax=axs[1, 0], label="Agri Land Area")
sns.lineplot(data=climate_df, x="year", y="cropland_pct", ax=axs[1, 0], label="Cropland %")
axs[1, 0].set_title("Agricultural Land Use Over Time")
axs[1, 0].legend()

# Fertilizer use and population density
sns.lineplot(data=climate_df, x="year", y="fertilizer_kg_per_ha", ax=axs[1, 1], label="Fertilizer Use")
sns.lineplot(data=climate_df, x="year", y="population_density", ax=axs[1, 1], label="Population Density")
axs[1, 1].set_title("Fertilizer Use and Population Density Over Time")
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Environment data analysis

```python
env_df = pd.read_csv(r'C:/Drive D/omdenaassignment/assignment7/capstone-project-sarvesh0427/data/environmentaldata/forest-area.csv')
```

```python
env_df.head()
```

```python
null(env_df)
```

```python
# Calculate forest land percentage
env_df["Forest %"] = (env_df["Forest land"] / env_df["Total Land"]) * 100

# Set style
sns.set(style="whitegrid")

# Plot: Top 10 districts by forest % coverage
top10 = env_df.sort_values("Forest %", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x="Forest %", y="District", palette="Greens_r")
plt.title("Top 10 Districts by Forest Coverage (%)")
plt.tight_layout()
plt.show()
```

```python
# Set style
sns.set(style="whitegrid")

# Plot: Top 10 districts by forest % coverage
top10 = env_df.sort_values("Forest %", ascending=True).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x="Forest %", y="District", palette="Reds_r")
plt.title("Top 10 Districts by less Forest Coverage (%)")
plt.tight_layout()
plt.show()
```
